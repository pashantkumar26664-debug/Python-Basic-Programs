```python
import matplotlib.pyplot as plt     
import pandas as pd
import numpy as np
```

Time → Line

Compare → Bar

Frequency → Histogram

Relationship → Scatter

Percentage → Pie

Outliers → Box

Total trend → Area

Pattern → Heatmap

```python
Days = [1,2,3,4,5]
Temp = [30,32,31,35,36]
plt.plot(Days, Temp)
plt.grid()
plt.show()
```

```python
Products = ['A','B','C']
Sales = [200,350,150]
plt.xlabel('Products')
plt.ylabel('Sales')
plt.bar(Products, Sales)
plt.show()
```

```python
Marks = [45,55,65,75,85,90,60,70]
plt.hist(Marks, bins=5)
# plt.grid()
plt.show()
```

```python
Hours = [1,2,3,4,5]
Marks = [40,50,65,70,85]
plt.scatter(Hours, Marks)
```

```python
Expense = [15,30,20,10]
Labels = ['Rent','Food','Travel','Others']
plt.pie(Expense, labels=Labels, autopct='%1.1f%%')
```

```python
Salaries = [30000,35000,40000,100000,45000]
plt.boxplot(Salaries)
```

```python

Days = [1, 2, 3, 4, 5]
Organic = [200, 250, 300, 280, 320]
Social = [100, 120, 140, 160, 180]
Ads = [50, 60, 70, 90, 100]

plt.stackplot(Days, Organic, Social, Ads,
              labels=['Organic', 'Social', 'Ads'])

plt.legend()
plt.xlabel("Days")
plt.ylabel("Visitors")
plt.title("Website Traffic Sources")
plt.show()
```

```python

attendance = np.array([
    [30, 28, 25, 27],
    [32, 30, 29, 31],
    [31, 29, 28, 30],
    [33, 31, 30, 32],
    [34, 32, 31, 33]
])

plt.imshow(attendance, cmap='hot')
plt.colorbar()
plt.xlabel("Periods")
plt.ylabel("Days")
plt.title("Student Attendance Heatmap")
plt.show()
```
