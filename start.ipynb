```python
a=([1,2,3],[4,5,6],[7,8,9])
ar=np.array(a)
ar
```

```python
ar.sum(axis=0)      #sum of element row wise
```

```python
ar.sum(axis=1)              #sum of all element column wise
```

```python
ar.T                #Transpose array
```

```python
ar.flat                      #print all element through flat
for item in ar.flat :
    print(item)
```

```python
ar.ndim          #find number of dimensional array
```

```python
ar.nbytes              #find how many bytes take by my array
```

```python
ar.argmax()              #find the index which maximun number in the array
```

```python
ar.argmax(axis=0)        #find the index which maximun number in the array in the axies
```

```python
ar.ravel()                #output:array([1, 2, 3, 4, 5, 6, 7, 8, 9])
```
